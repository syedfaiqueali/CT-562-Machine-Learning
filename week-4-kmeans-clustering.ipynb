```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D

import warnings
warnings.filterwarnings('ignore')
```

```python
data = np.array([[61, 31],
                 [59, 35],
                 [18, 35],
                 [15, 34],
                 [82, 52],
                 [21, 30],
                 [23, 41],
                 [65, 36],
                 [56, 33]])

# Define KMeans model with 2 clusters
kmeans = KMeans(n_clusters=2, random_state=42)

# Fit the model to the data
kmeans.fit(data)

# Get cluster labels
cluster_labels = kmeans.labels_

# Define labels for clusters
cluster_labels_dict = {0: 'human', 1: 'car'}

# Assign labels to clusters
cluster_labels_mapped = [cluster_labels_dict[label] for label in cluster_labels]

# Display cluster labels
for i, label in enumerate(cluster_labels_mapped):
    print(f"Data point {i+1} belongs to cluster: {label}")
```

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

for i, point in enumerate(data):
    if cluster_labels_mapped[i] == 'human':
        ax.scatter(point[0], point[1], 0, c='blue', label='Human')
    else:
        ax.scatter(point[0], point[1], 0, c='red', label='Car')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Scatter Plot of KMeans Clustering')

plt.show()
```
